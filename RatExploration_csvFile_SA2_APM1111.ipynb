```python
import pandas as pd
from statsmodels.formula.api import ols
import statsmodels.api as sm
from tabulate import tabulate

id_vars = ['Stimuli']
value_vars = ['Time']

data_analyzed = pd.melt(df, id_vars=id_vars, value_vars=value_vars, var_name='Condition', value_name='Seconds')

formula = 'Seconds ~ Stimuli'
analyzed_frame = ols(formula, data=data_analyzed).fit()
anova_df = sm.stats.anova_lm(analyzed_frame, typ=2)

print(tabulate(anova_df, headers='keys', tablefmt='pretty'))
```

# Interpretation:

 
<h5>

<li> Stimuli Group <br>
Sum of Squares (SS): 44.527: This illustrates the overall variation in the amount of time that each stimulus group spent exploring the chamber.
<br>
Degrees of Freedom (df): 2: There are two degrees of freedom associated with the stimuli group, indicating the number of independent comparisons being made.<br>

Mean Square (MS): 22.263: This represents the average variation in the amount of time that each stimulus group spent exploring the chamber.
<br>

<br><br>sss

<li> Residuals <br>
Sum of Squares (SS): 11.833: This represents the residual variability in the data after accounting the differences between stimuli.<br>

Degrees of Freedom (df): 33: The residuals show 33 degrees of freedom, which represents the variability that the stimuli group was unable to account for.<br>

Mean Square (MS): 0.359: This is the average unexplained variability in the time spent exploring the chamber.<br>



Since F (62.089) with greater variation between group means is indicated by a bigger F-ratio, is a metric that represents the ratio of variance within groups to variance between groups.
<br><br><br><br>
<b>Additional, pairs of comparisons could be carried out in light of the statistically significant results to determine which particular stimuli groups differ from one another significantly.
